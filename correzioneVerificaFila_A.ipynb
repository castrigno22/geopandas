10)il numero di consolati in ogni continente

```python
result = db.consolati.aggregate([{"$group" :{"_id":"$continente", "total" : {"$sum" : 1}}},
                                 {"$sort" :{"_id" : 1}}])
df = pd.DataFrame(list(result))
df
```

11)i continenti che hanno piÃ¹ di 50 consolati. Ordinare in ordine decrescente a partire dal continente che ha piÃ¹ consolati


```python
result = db.consolati.aggregate([
                                 {"$group" :{"_id":"$continente", "total" : {"$sum" : 1}}},
                                 {"$match" :{"total":{"$gt":50}}},
                                 {"$sort" :{"total" : -1}}])
df = pd.DataFrame(list(result))
df
```
