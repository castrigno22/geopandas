4. Visualizzare la provincia di Milano su una mappa di contextily con i confini amministrativi del comune di Milano. Usare colori adatti

```python
milano = province.loc[province['DEN_CM'] == 'Milano', 'geometry'].squeeze()
```

```python
comuni_milano = comuni[comuni.within(milano)]
ax = comuni_milano.to_crs(epsg= 3857).plot(figsize= (10, 12), facecolor = "none", edgecolor = "b")
contextily.add_basemap(ax, source=contextily.providers.OpenStreetMap.Mapnik)
```

5. Utilizzando le informazioni geografiche riguardanti i quartieri di Milano, visualizzare sia i confini di Milano sia i confini dei quartieri.

```python
quartieri_milano = geopandas.read_file('/content/drive/MyDrive/Colab Notebooks/Geopandas - Materiale esercizi/ds964_nil_wm.zip')
```

```python
ax = comuni_milano.to_crs(epsg=3857).plot(figsize= (10, 12), facecolor = "none", edgecolor = "b")
quartieri_milano.to_crs(epsg=3857).plot(ax=ax, facecolor = "none", edgecolor= "red")
contextily.add_basemap(ax, source=contextily.providers.OpenStreetMap.Mapnik)
```

6. Far inserire all'utente una provincia della Lombardia e visualizzare tutte le province lombarde  confinanti. Controllare che la provincia inserita sia effettivamente in Lombardia.

```python
prov = input('Inserire una provincia della Lombardia: ')
provincia = prov.capitalize()
provincia1 = province[province["DEN_UTS"] == provincia].geometry.squeeze()
if provincia1.within(lombardia) == True: 
  ax = province[province.touches(provincia1) & province.within(lombardia)].to_crs(epsg=3857).plot(figsize=(12, 10), alpha = 0.3 , edgecolor = "k")
  contextily.add_basemap(ax, source=contextily.providers.OpenStreetMap.Mapnik)
else:
  print("La provincia non si trova in Lombardia")
```
