```python
import pandas as pd
import geopandas 
import contextily
quartieri = gpd.read_file("/content/drive/MyDrive/Colab Notebooks/ds964_nil_wm.zip")
quartieri.head(13)
```

```python
quartieri.crs
```

```python
from shapely.geometry import Point
```

```python
duomo=Point(9.190364,45.463938)
```

```python
ax = quartieri[quartieri.contains(duomo)].to_crs(epsg=3857).plot(figsize=(10,8),facecolor="none",edgecolor="k")
geopandas.GeoSeries([duomo],crs="EPSG:4326").to_crs(epsg=3857).plot(ax=ax,color="red")
contextily.add_basemap(ax)
```

```python
print("7.visualizzare i quartieri vicini a quello che contiene il mondo")
ViciniDuomo=quartieri[quartieri.touches(quartieri[quartieri.contains(duomo)].geometry.squeeze())]
ax=ViciniDuomo.to_crs(epsg=3857).plot(figsize=(10,8),facecolor="none",edgecolor="k")
contextily.add_basemap(ax)
```
